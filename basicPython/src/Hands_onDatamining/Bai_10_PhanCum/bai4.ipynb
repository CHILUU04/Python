```python
data = df.copy()
data['KhoiA'].fillna(data['KhoiA'].mean(), inplace=True)
data['KhoiB'].fillna(data['KhoiB'].mean(), inplace=True)
data['KhoiC'].fillna(data['KhoiC'].mean(), inplace=True)
data['KhoiD'].fillna(data['KhoiD'].mean(), inplace=True)
data['KhoiA1'].fillna(data['KhoiA1'].mean(), inplace=True)
data
# data['KhoiA'].isnull().sum()
```

```python
#xem sự phân bô diem giưa các khoi
plt.boxplot((data['KhoiA'], data['KhoiA1'], data['KhoiB'], data['KhoiC'], data['KhoiD']),
            label=['Khối A', 'Khối A1', 'Khối B', 'Khối C', 'Khối D'])
plt.title('Sự phân bô điểm')
plt.show
```

```python

# Danh sách tên cột của 5 khối


cols = ['KhoiA','KhoiA1','KhoiB','KhoiC','KhoiD']

for col in cols:
    # Tính Q1, Q3 và IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Xác định ngưỡng
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # “Capping” (hay còn gọi là winsorization) là phương pháp giới hạn (thay vì loại bỏ) 
    # các giá trị nằm ngoài ngưỡng outlier về đúng giá trị biên tương ứng:
    
      # Capping:
    data[col] = np.where(data[col] < lower, lower,
                         np.where(data[col] > upper, upper, data[col]))

# Vẽ boxplot
plt.boxplot([data[c] for c in cols], labels=cols)
plt.title('Phân bố điểm các khối (loại outlier theo IQR)')
plt.ylabel('Điểm')
plt.grid(True)
plt.show()
```
