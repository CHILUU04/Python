```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist # hỗ trợ tính khoảng cách
```

```python
mean = [[2, 2], [9, 2], [4, 9], [8, 2], [9, 8]]
cov = [[2, 0], [0, 2]]
n_sample = 500 # số mẫu mỗi cụm
n_clutes = 5
x0 = np.random.multivariate_normal(mean[0], cov, n_sample)
x1 = np.random.multivariate_normal(mean[1], cov, n_sample)
x2 = np.random.multivariate_normal(mean[2], cov, n_sample)
x3 = np.random.multivariate_normal(mean[3], cov, n_sample)
x4 = np.random.multivariate_normal(mean[4], cov, n_sample)
X = np.concatenate((x0, x1, x2, x3, x4), axis=0)
x4
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'bo', markersize = 5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #  vẽ dữ liệu trong không gian 3 chiều
from termcolor import colored as cl # để in màu hoặc định dạng text 

# from sklearn.preprocessing import StandardScaler # data normalization
from sklearn.cluster import KMeans # K-means algorithm

clusters = 4
model = KMeans(
    n_clusters = clusters
)
model.fit(X)
lables = model.labels_
print(cl(lables[:100], attrs = ['bold']))
```

```python
model = KMeans(
                init= 'k-means++', ## cách chọn tâm khởi tạo thông minh
                n_clusters=clusters,  # số cụm (k)
                n_init=20)   # số lần khởi tạo ngẫu nhiên rồi chọn kết quả tốt nhất
```

```python
model.fit(X)
labels = model.labels_
print(cl(labels[:100], attrs=['bold']))
```
