```python
import pandas as pd
import sklearn
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df = pd.read_csv('D:\\learn_PyThon\\basicPython\\src\\Bai_4+5\\Dlht.csv')
# df
```

```python
df = pd.get_dummies(df, columns=['Gender'], drop_first=True, dtype=int)
df
```

```python

# Chuyển đổi cột skill thành danh sách các kỹ năng
data1 = df.copy();
data1 = data1.join(data1['Skill'].str.get_dummies(sep=','))
data1 = data1.drop(columns=['Skill']) 
# data1
# data1['Skill'].str.get_dummies(sep=','):
# Hàm str.get_dummies(sep=',') hoạt động trên chuỗi ký tự trong cột "Skill". Nó thực hiện các bước sau:
# Tách các giá trị trong chuỗi thành các phần tử riêng biệt theo dấu phẩy (sep=',').
# Tạo ra các cột mới, mỗi cột đại diện cho một kỹ năng duy nhất có trong cột "Skill".
# Ở mỗi hàng, nếu kỹ năng đó có mặt thì giá trị ở cột tương ứng là 1, nếu không có thì là 0.
# .join(...):
# Sau khi tạo ra DataFrame chứa các dummy variables, hàm join sẽ nối các cột này vào DataFrame ban đầu (data1). Như vậy, DataFrame mới sẽ có thêm các cột tương ứng với từng kỹ năng.
```

```python
# Rời rạc hóa trường Salary thành 3 mức: 'Low','Medium','High'
cut_lables = ['Low','Medium','High']
cut_bin = [0, 800, 1200, 2000]
data1['sat_cat'] = pd.cut(data1['Salary'], bins=cut_bin, labels=cut_lables) #bins: là danh sách các điểm giá trị trong khoản cần cắt và lables: để gán tên cho khoản (bins)
# print(data1['sat_cat'])
# print(data1)
```

```python
# Chuyển trường OverseaProject về dạng 0, 1 với {'Yes':1, 'No':0}
#ta có thể dùng map: nó giúp với mỗi giá trị gốc, nó sẽ tra và thay thế bằng giá trị mới theo quy tắc bạn định nghĩa.
df['OverseaProject'] = df['OverseaProject'].map({'Yes': 1, 'No': 0})
df
```

```python
# Xóa trường Gender, OverseaProject
data1 = df.copy()
df_new = data1.drop(['Gender', 'OverseaProject'], axis=1) # axis = 1là xóa theo cột và ngược lại
df_new
```
