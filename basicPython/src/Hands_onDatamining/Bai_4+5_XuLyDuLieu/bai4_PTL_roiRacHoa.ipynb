```python
# 1. Lấy danh sách cột numeric
num_cols = df.select_dtypes(include='number').columns

# 2. Chuyển từng cột numeric đều về 0/1
df[num_cols] = (df[num_cols] >= 6).astype(int)

# 3. Ghi ra file
df.to_csv('105moi.csv', index=False)
```

```python
# 2. Xác định các cột điểm (chỉ các cột numeric)
score_cols = df.select_dtypes(include='number').columns.tolist()
#df.select_dtypes(include='number') : Hàm này trả về một DataFrame con chỉ gồm các cột có kiểu số (int, float…) từ df.
#.columns: Trên DataFrame kết quả ở bước 1, .columns lấy ra Index (đối tượng giống như list) chứa tên các cột, ví dụ Index(['A','C'], dtype='object').
#.tolist(): Chuyển đối tượng Index đó thành list Python thực thụ, tức ['A','C'].

# 3. Chuyển mỗi ô thành True nếu có điểm (không phải NaN), False nếu thiếu điểm
df[score_cols] = df[score_cols].notna()
#với notna thi .notna() trả về True cho ô không phải NaN (có điểm), False cho ô NaN (không có điểm).
# 4. Lưu kết quả ra file mới
df.to_csv('105OK.csv', index=False)

print("Đã tạo file 105OK.csv")
```
