```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
data = pd.read_csv('C:\\Users\\Minh Chi\\Downloads\\boDuLieu\\lap 6\\105.csv')
data
```

```python
data.head(10)
```

```python
records = []
for i in range(90):
    row = [str(data.values[i, j])
            for j in range(10)]
    records.append(row)

association_rules = apriori(
    records,
    min_support=0.01,       
    min_confidence=0.3,    
    min_lift=3, #lift đo xem A và B cùng xuất hiện cao hơn bao nhiêu lần so với khi chúng độc lập.
    #giúp bạn loại bỏ các luật dù có confidence ổn nhưng không có mối liên hệ thật sự mạnh so với ngẫu nhiên
    min_length=2
)

association_results = list(association_rules) 
#  để “kéo” toàn bộ luật ra vào một danh sách association_results, để ta có thể lặp hoặc truy cập nhiều lần.
print(association_results)
for item in association_results:
#         Mỗi item ở đây là một tuple có dạng:
            # (
            #   item[0],        # tập item (frequent itemset), ví dụ frozenset({'A','C'})
            #   item[1],        # support của tập này, ví dụ 0.5
            #   item[2]         # list các OrderedStatistic, chứa confidence & lift
            # )
    pair = item[0]
    items = [x for x in pair]
    print("Rule : " + items[0] +  " -> " + items[1])
    print("spport : " + str(item[1]))
    print("confidence : " + str(item[2][0][2]))
    # item[2] là một list các OrderedStatistic. Mỗi OrderedStatistic chứa các giá trị:
    # items_base (tiền đề),
    # items_add (hậu đề),
    # confidence,
    # lift.
    print("Lift : "+ str(item[2][0][3]))
    print("--------------------------------------------")
    
```
