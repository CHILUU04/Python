```python
# #chuyen data ve dang 0 1
df_ohe = df.notnull().astype(int) # nhìn dữ liệu thay vì dùng filler cụ thể missing rồi one-hot phức tạp thì nên dung 
# #cụ thể đây là dữ liệu có mỗi tên mặt hàng
# # nếu khác mua thì có tên k mua thì NaN
df_ohe

```

```python
df_ohe.shape
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
#tim tập xuất hiện thường xuyên với min_support = 50%
frenquent_items = apriori(df_ohe, min_support=0.5, use_colnames=True)
frenquent_items
```

```python
#sinh luật kết hợp với độ tin cậy 80%   
rules = association_rules(frenquent_items, metric='confidence', min_threshold= 0.8)
rules
```

```python
rules[(rules['lift'] > 1) & 
      (rules['confidence'] >= 0.7)].sort_values(['confidence'], ascending = False)
```

```python
# làm đối với fb-growth
from mlxtend.frequent_patterns import fpgrowth, association_rules

#làm tương tự với dộ hộ trợ và tin cậy giống trên 40% và 70%
frenquent_items_fp = fpgrowth(df_ohe, min_support=0.4, use_colnames=True)
frenquent_items_fp
```

```python
rules_fp = association_rules(frenquent_items_fp, metric='confidence', min_threshold=0.8)
rules_fp
```
