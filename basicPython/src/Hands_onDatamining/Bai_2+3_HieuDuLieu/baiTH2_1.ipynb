```python
# Tính tổng số bản ghi trong DataFrame
total_records = len(df)

# Thống kê số lượng các tổ hợp giá trị duy nhất cho các cột phân loại
category_stats = (
    df[['Dept', 'Gender', 'Skill', 'OverseaProject']]
    .value_counts()
    .rename_axis(['Dept', 'Gender', 'Skill', 'OverseaProject']) #giúp đặt lại tên trục chỉ mục cho các cột này.
    .reset_index(name='Count') #  tên cột Count cho giá trị đếm được.
)

# Tính tỷ lệ phần trăm của mỗi tổ hợp so với tổng số bản ghi
category_stats['Percentage'] = (category_stats['Count'] / total_records) * 100
print(category_stats['Percentage'])
```
