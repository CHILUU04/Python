```python
from pandas.api.types import is_numeric_dtype
cols = ['Age', 'WorkingYear', 'TrainedYear', 'Dept', 'Gender']
for col in cols:
    if is_numeric_dtype(df[col]):
        df[col].fillna(df[col].mean(), inplace=True)
    else : 
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['Gender_binary'] =  df['Gender'].map({'Male': 1, 
                                'Female': 0})
df.drop('Gender', axis=1)
```

```python
#chuan hóa về dạng 0 1 sử dụng one-hot encoding
# df['Skill']
df1 = df.copy()
df1 = pd.get_dummies(df1['Skill'], drop_first=True, dtype=int)

df = pd.concat([df, df1], axis=1)
df.head(5)
```

```python
df.drop('Skill', axis=1, inplace=True)
```

```python
cut_labels = ['Low', 'medium', 'high']
cut_bins = [0, 100, 1000, 10000]

df['Salary'] = pd.cut(df['Salary'], bins=cut_bins, labels=cut_labels)
df['Salary'].head(10)
```

```python

df['OverseaProject_binary']= df['OverseaProject'].map({'Yes': 1, 'No': 0})
df.head(10)
```

```python
df.drop('OverseaProject', axis= 1, inplace=True)
df.head(5)
```

```python
df['TrainedYear'].info()
```

```python
cut_labels = ['College', 'Univesity', 'Mater_PhD']
cut_bins = [0, 2, 4, 8]
df['TrainedYear'] = pd.cut(df['TrainedYear'], bins = cut_bins, labels = cut_labels)
```

```python
data.drop('ID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
scaled = mms.fit_transform(data)
```

```python
data_mms = pd.DataFrame(scaled, columns=['WorkingYears', 'Salary'])
```

```python

plt.boxplot(
    (data_mms['Salary'], data_mms['WorkingYears']),
    labels=['Salary', 'WorkingYears']
)
plt.title('Boxplot of Salary and WorkingYears (Scaled)')
plt.ylabel('Scaled Value')
plt.show()
```

```python
plt.scatter(data_mms['WorkingYears'], data_mms['Salary'],s = 150 ,color = 'skyblue', edgecolor = 'b')
plt.title('Scatter plot of workingYears and Salary')
plt.xlabel('Working Year')
plt.ylabel('Salary')
plt.show()
```
