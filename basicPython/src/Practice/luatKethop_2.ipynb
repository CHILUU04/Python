```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from sklearn.preprocessing import OneHotEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('C:\\Users\\Minh Chi\\Downloads\\boDuLieu\\lap 6\\Market_Basket_AR.csv')
df.head(5)
```

```python
df.info()
```

```python
#chuẩn hóa one-hot encoding

# 2) Chọn các cột cần one-hot (ví dụ: object hoặc category)
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
# Dùng scikit-learn OneHotEncoder (dễ tích hợp pipeline) 
encoder  = OneHotEncoder(sparse_output=False)
# Fit & transform chỉ trên cột phân loại
one_hot_encoded  = encoder.fit_transform(df[cat_cols])

# Tạo DataFrame one-hot với tên cột rõ ràng
one_hot_df  = pd.DataFrame(
    one_hot_encoded,
    columns=encoder.get_feature_names_out(cat_cols),
    index=df.index
)

# Nối lại với các cột số nguyên, số thực ban đầu và xoá cột gốc
df_encoded = pd.concat([df, one_hot_df], axis=1)

df_encoded = df_encoded.drop(cat_cols, axis=1)
df_encoded.head(10)
df_encoded['shrimp_almonds']
```

```python
df_encoded.drop('olive oil', axis=1, inplace=True)
```

```python
frequren_items = apriori(df_encoded, min_support=0.7, use_colnames=True)
frequren_items
```

```python
rules = association_rules(frequren_items, metric='confidence', min_threshold=0.7)
rules
```
